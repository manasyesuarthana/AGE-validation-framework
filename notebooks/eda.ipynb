```python
import pandas as pd
import numpy as np
from scipy.io import arff

def load_arff_to_dataframe(file_path):
    """
    Load an ARFF file and convert it to a pandas DataFrame.

    Parameters:
    file_path (str): The path to the ARFF file.

    Returns:
    pd.DataFrame: A DataFrame containing the data from the ARFF file.
    """
    # Load the ARFF file
    data = arff.loadarff(file_path)
    
    # Convert to pandas DataFrame
    df = pd.DataFrame(data[0])
    
    # Decode byte strings to regular strings if necessary
    for col in df.select_dtypes([object]):
        df[col] = df[col].apply(lambda x: x.decode('utf-8') if isinstance(x, bytes) else x)
    
    return df
```

```python
# Load all ARFF files
PATH = '/home/Projects/AGE-validation-framework/sim/data'
train_d1 = load_arff_to_dataframe(f'{PATH}/EpilepsyDimension1_TRAIN.arff')
test_d1 = load_arff_to_dataframe(f'{PATH}/EpilepsyDimension1_TEST.arff')
train_d2 = load_arff_to_dataframe(f'{PATH}/EpilepsyDimension2_TRAIN.arff')
test_d2 = load_arff_to_dataframe(f'{PATH}/EpilepsyDimension2_TEST.arff')
train_d3 = load_arff_to_dataframe(f'{PATH}/EpilepsyDimension3_TRAIN.arff')
test_d3 = load_arff_to_dataframe(f'{PATH}/EpilepsyDimension3_TEST.arff')

# Display shapes
print("Dimension 1 - Train:", train_d1.shape, "Test:", test_d1.shape)
print("Dimension 2 - Train:", train_d2.shape, "Test:", test_d2.shape)
print("Dimension 3 - Train:", train_d3.shape, "Test:", test_d3.shape)
```

```python
def merge_datasets(train_df1, train_df2, train_df3, test_df1, test_df2, test_df3):
    """
    Merge training and testing datasets from three dimensions.

    Parameters:
    train_df1, train_df2, train_df3 (pd.DataFrame): Training DataFrames for dimensions 1, 2, and 3.
    test_df1, test_df2, test_df3 (pd.DataFrame): Testing DataFrames for dimensions 1, 2, and 3.

    Returns:
    pd.DataFrame, pd.DataFrame: Merged training and testing DataFrames.
    """
    # Merge training datasets
    train_merged = pd.concat([train_df1.reset_index(drop=True), 
                              train_df2.reset_index(drop=True), 
                              train_df3.reset_index(drop=True)], axis=1)
    
    # Merge testing datasets
    test_merged = pd.concat([test_df1.reset_index(drop=True), 
                             test_df2.reset_index(drop=True), 
                             test_df3.reset_index(drop=True)], axis=1)
    
    return train_merged, test_merged
```

```python
train_merged, test_merged = merge_datasets(train_d1, train_d2, train_d3, test_d1, test_d2, test_d3)
```
